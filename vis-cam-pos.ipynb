```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def plot_cam_positions_with_ids(cam_pos_list):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    for cam_id, pos in cam_pos_list:
        x, y, z = pos
        ax.scatter(x, y, z, c='red', s=50)
        ax.text(x, y, z, str(cam_id), fontsize=10, color='black', ha='center', va='bottom')

    ax.set_title("Camera Positions with IDs")
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("Z")
    ax.grid(True)
    plt.tight_layout()
    plt.show()


plot_cam_positions_with_ids(cam_pos_discussion)
```
